## 고궁/전통문화체험 모두 참여한 사람들 중, 고궁 또는 전통문화체험을 ‘만족한 활동’ 선택 여부

```python
import pandas as pd


# 참여활동 관련 컬럼 리스트
participation_cols = [f'Q8a{str(i).zfill(2)}' for i in range(1, 21)]
# 만족활동 관련 컬럼 리스트
satisfaction_cols = [f'Q8_1a{i}' for i in range(1, 4)]

# 고궁/전통문화체험 모두 참여한 사람 필터링
def has_participated(row):
    participated_5 = any(row[col] == 5 for col in participation_cols)
    participated_6 = any(row[col] == 6 for col in participation_cols)
    return participated_5 and participated_6

df_2024['both_participated'] = df_2024.apply(has_participated, axis=1)

# 만족활동 중 고궁(5) 또는 전통문화체험(6)을 선택했는지 여부
def satisfied_with_either(row):
    return any(row[col] in [5, 6] for col in satisfaction_cols)

df_2024['satisfied_with_either'] = df_2024.apply(satisfied_with_either, axis=1)

# 두 가지 모두 참여한 사람 중 만족 활동에 하나라도 포함된 비율
subset = df_2024[df_2024['both_participated']]
satisfied_count = subset['satisfied_with_either'].sum()
total_count = len(subset)
satisfaction_rate = satisfied_count / total_count if total_count > 0 else 0

print(f"총 인원: {total_count}")
print(f"만족 활동에 고궁/전통문화 포함한 인원: {satisfied_count}")
print(f"비율: {satisfaction_rate:.2%}")
```

## 만족한 사람들 분류

| 그룹명       | 만족 활동 내용                |
| --------- | ----------------------- |
| A. 둘 다 만족 | 고궁(5), 전통문화체험(6) 모두 선택  |
| B. 고궁만 만족 | 고궁(5)만 있고 전통문화체험(6)은 없음 |
| C. 전통만 만족 | 전통문화체험(6)만 있고 고궁(5)은 없음 |
| D. 둘 다 없음 | 고궁(5), 전통문화체험(6) 모두 없음  |


```python
# 참여 조건: 고궁(5), 전통(6) 모두 참여한 사람
def has_participated_both(row):
    return any(row[col] == 5 for col in participation_cols) and \
           any(row[col] == 6 for col in participation_cols)

# 만족 활동 선택 여부
def get_satisfaction_group(row):
    sats = [row[col] for col in satisfaction_cols]
    has_5 = 5 in sats
    has_6 = 6 in sats

    if has_5 and has_6:
        return 'A. 고궁+전통 모두 만족'
    elif has_5 and not has_6:
        return 'B. 고궁만 만족'
    elif has_6 and not has_5:
        return 'C. 전통만 만족'
    else:
        return 'D. 둘 다 없음'

# 필터링 및 분류
df_2024['both_participated'] = df_2024.apply(has_participated_both, axis=1)
df_2024_participants = df_2024[df_2024['both_participated']].copy()
df_2024_participants['satisfaction_group'] = df_2024_participants.apply(get_satisfaction_group, axis=1)

# 비율 집계
group_counts = df_2024_participants['satisfaction_group'].value_counts().sort_index()
group_ratios = (group_counts / group_counts.sum()).sort_index()

# 출력
print("각 만족 그룹별 인원 수:\n", group_counts)
print("\n각 만족 그룹별 비율:\n", group_ratios.apply(lambda x: f"{x:.2%}"))
```

```python
import matplotlib.pyplot as plt

# 시리즈 정렬 (앞서 계산한 group_ratios가 있다고 가정)
group_ratios = group_ratios.sort_index()

# 막대그래프 그리기
plt.figure(figsize=(8, 5))
plt.bar(group_ratios.index, group_ratios.values, color='skyblue')
plt.title('고궁+전통문화체험 참여자 만족 활동 분포')
plt.ylabel('비율')
plt.ylim(0, 1)

# 막대 위에 비율 표시
for i, v in enumerate(group_ratios.values):
    plt.text(i, v + 0.02, f"{v:.1%}", ha='center', fontsize=10)

plt.xticks()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 만족 활동 컬럼
satisfaction_cols = ['Q8_1a1', 'Q8_1a2', 'Q8_1a3']

# 고궁(5), 전통문화체험(6) 포함 여부 판단 함수
def classify_satisfaction(row):
    values = set(row[col] for col in satisfaction_cols if pd.notna(row[col]))
    has_5 = 5 in values
    has_6 = 6 in values

    if has_5 and has_6:
        return 'A. 고궁+전통 모두 만족'
    elif has_5:
        return 'B. 고궁만 만족'
    elif has_6:
        return 'C. 전통만 만족'
    else:
        return 'D. 둘 다 없음'

# 분류 컬럼 추가
df_2024['satisfaction_group'] = df_2024.apply(classify_satisfaction, axis=1)

# 그룹별로 분할
df_2024_both = df_2024[df_2024['satisfaction_group'] == 'A. 고궁+전통 모두 만족']
df_2024_only_palace = df_2024[df_2024['satisfaction_group'] == 'B. 고궁만 만족']
df_2024_only_traditional = df_2024[df_2024['satisfaction_group'] == 'C. 전통만 만족']
df_2024_neither = df_2024[df_2024['satisfaction_group'] == 'D. 둘 다 없음']
```

## 각 그룹의 좋았던 장소 TOP20

## 혼자 온 외래관광객 중 활동별로 재방문 의향

(긍정 응답 비율: Q13이 1 또는 2)을 분석

- 혼자 온 사람 필터 (Q7A == 1)

- 주요 활동 중 하나라도 참여한 경우 그룹핑

- 각 활동별로 Q13이 4 또는 5(긍정적 응답)인 비율 계산

- 재방문 의향이 높은 활동 TOP N 출력

```python
import pandas as pd

# 대상: 혼자 온 외래관광객
solo_df = df_2024[df_2024['Q7A'] == 1].copy()

# 주요 활동 선택 항목 (응답자는 Q8a01 ~ Q8a20에 '1~20 중 코드 번호'를 채워 넣음)
q8a_cols = [f'Q8a{str(i).zfill(2)}' for i in range(1, 21)]

# Q13 숫자형으로 변환
solo_df['Q13'] = pd.to_numeric(solo_df['Q13'], errors='coerce')

# 활동 코드 ↔ 활동명 매핑
activity_map = {
    1: '식도락(음식/미식)',
    2: '쇼핑',
    3: '자연경관 감상',
    4: '휴양/웰니스',
    5: '고궁/역사 유적지 방문',
    6: '전통문화체험',
    7: '박물관/전시관 관람',
    8: 'K-POP/한류스타/드라마 촬영지',
    9: '공연 관람 (뮤지컬 등)',
    10: '지역 축제 참여',
    11: '유흥/카지노',
    12: '놀이공원/테마파크',
    13: '뷰티/미용 관광',
    14: '치료/건강검진',
    15: '스포츠 관람',
    16: '스포츠 참가',
    17: '국제회의/전시',
    18: '비즈니스 활동',
    19: '연수/교육/연구',
    20: '기타'
}

# 결과 저장 리스트
results = []

# 활동 코드 1~20에 대해 반복
for code in range(1, 21):
    # 해당 코드가 Q8a01~Q8a20 중 어느 칼럼에서든 존재하는 사람 필터링
    mask = solo_df[q8a_cols].isin([code]).any(axis=1)
    participants = solo_df[mask]
    total = len(participants)

    if total == 0:
        continue

    # 재방문 의향자 (Q13 == 4 or 5)
    positive = participants[participants['Q13'].isin([4, 5])]
    positive_count = len(positive)
    positive_ratio = round(positive_count / total * 100, 2)

    results.append({
        '활동코드': code,
        '활동명': activity_map[code],
        '참여자 수': total,
        '재방문 의향 있음': positive_count,
        '재방문 비율(%)': positive_ratio
    })

# 정렬 및 출력
result_df = pd.DataFrame(results)
result_df = result_df.sort_values(by='재방문 비율(%)', ascending=False).reset_index(drop=True)
result_df
```
