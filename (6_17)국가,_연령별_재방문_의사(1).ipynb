## 2019년

#### **1. 국가, 연령별 3년내 재방문 의사 확인**

```python
country_map = {
    1: '일본',
    2: '중국',
    3: '홍콩',
    4: '싱가포르',
    5: '대만',
    6: '태국',
    7: '말레이시아',
    8: '호주',
    9: '미국',
    10: '캐나다',
    11: '영국',
    12: '독일',
    13: '프랑스',
    14: '러시아',
    15: '중동',
    16: '인도',
    17: '필리핀',
    18: '인도네시아',
    19: '베트남',
    20: '몽골',
    97: '기타'
}

tour_2019['국가명'] = tour_2019['D_COU1'].map(country_map)
```

[주요변수]
- Q16 : 재방문의사(1~5점)
- D_AGE : 연령대
- D_COU1 : 국적

**국가별 방문 횟수**

```python
tour_2019.groupby("국가명")["Q1"].count().sort_values(ascending=False)
```

**연령대별 재방문 의향 평균**

```python
# 재방문 의향 평균 (연령대별)
tour_2019.groupby("D_AGE")["Q16"].mean().sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 박스플롯
sns.boxplot(data=tour_2019, x="D_AGE", y="Q16")
plt.title("연령대별 향후 3년 내 재방문 의향 분포")
plt.show()
```

**국적별 재방문 의향 평균**

```python
# 재방문 의향 평균 (국적별)
tour_2019.groupby("국가명")["Q16"].mean().sort_values(ascending=False)
```

**연령대x국적별 평균 재방문 의향**

```python
# 연령대 × 국적별 평균 재방문 의향
tour_2019.groupby(["국가명", "D_AGE"])["Q16"].mean().unstack().round(2)
```

**국가별 재방문 4~5점 비율**

```python
# 국가별 4~5점 응답자 비율
tour_2019['high_revisit'] = tour_2019['Q16'] >= 4
tour_2019.groupby('국가명')['high_revisit'].mean().sort_values(ascending=False)
```
